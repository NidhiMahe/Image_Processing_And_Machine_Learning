```python
from skimage import io
from skimage import draw
img = io.imread('dog.jpg')
def rectangle(x, y, w, h):
    rr, cc = [x, x + w, x + w, x], [y, y, y + h, y + h]
    return (draw.polygon(rr, cc))

rr, cc = rectangle(50, 115, 70,70)
img[rr, cc] = 1
io.imshow(img)

```

```python
from skimage import io
from skimage import draw

# Load the image
img = io.imread('dog.jpg')

# Draw a bold line by drawing multiple parallel lines
thickness = 5  # Adjust the thickness here

for i in range(thickness):
    x, y = draw.line(0 + i, 0, 150 + i, 200)
    img[x, y] = 0  # Set the line color to black (0)

# Display the image with the bold line
io.imshow(img)
io.show()
```

```python
import cv2
from skimage import io

# Load the image
img = io.imread('dog.jpg')

# Convert the image from RGB (skimage) to BGR (OpenCV)
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

# Specify the text and position
text = "Puppy"
position = (50, 50)  # (x, y) coordinates

# Choose font, size, color, and thickness
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1
color = (255, 0, 0)  # Blue color in BGR
thickness = 2

# Add text to the image
cv2.putText(img, text, position, font, font_scale, color, thickness)
```

```python
io.imshow(img)
```
