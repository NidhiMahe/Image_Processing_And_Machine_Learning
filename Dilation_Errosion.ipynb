```python
'''Erosion leads to the removal of pixels from the boundary.
Dilation results in the addition of pixels to the boundary of an object; '''
```

```python
#Import package
import cv2
```

```python
from skimage import io
```

```python
#Read image
image = cv2.imread("/users/Nidhi/Desktop/dog.jpg")
```

```python
io.imshow(image)
```

```python
'''OpenCV uses the BGR (Blue, Green, Red) color format by default, while most other libraries, including scikit-image, use the RGB 
(Red, Green, Blue) format. As a result, when you read and display an image with OpenCV, it may appear with incorrect colors if you do 
not convert it to the RGB format.'''
```

```python
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
io.imshow(img_rgb)
```

```python
#Define erosion size
s1 = 0
s2 = 10
s3 = 10
```

```python
#Define erosion type
t1 = cv2.MORPH_RECT
t2 = cv2.MORPH_CROSS
t3 = cv2.MORPH_ELLIPSE
```

```python
#Define and save the erosion template
tmp1 = cv2.getStructuringElement(t1, (2*s1 + 1, 2*s1+1), (s1, s1))
tmp2= cv2.getStructuringElement(t2, (2*s2 + 1, 2*s2+1), (s2, s2))
tmp3 = cv2.getStructuringElement(t3, (2*s3 + 1, 2*s3+1), (s3, s3))
```

```python
#Apply the erosion template to the image and save in different variables
final1 = cv2.erode(img_rgb, tmp1)
final2 = cv2.erode(img_rgb, tmp2)
final3 = cv2.erode(img_rgb, tmp3)
```

```python
#Show all the images with different erosions
io.imshow(final1)
```

```python
io.imshow(final2)
```

```python
io.imshow(final3)
```

```python
#Define dilation size
d1 = 0
d2 = 10
d3 = 20
```

```python
#Define dilation type
t1 = cv2.MORPH_RECT
t2 = cv2.MORPH_CROSS
t3 = cv2.MORPH_ELLIPSE
```

```python
#Store the dilation templates
tmp1 = cv2.getStructuringElement(t1, (2*d1 + 1, 2*d1+1), (d1, d1))
tmp2 = cv2.getStructuringElement(t2, (2*d2 + 1, 2*d2+1), (d2, d2))
tmp3 = cv2.getStructuringElement(t3, (2*d3 + 1, 2*d3+1), (d3, d3))
```

```python
#Apply dilation to the images
final1 = cv2.dilate(img_rgb, tmp1)
final2 = cv2.dilate(img_rgb, tmp2)
final3 = cv2.dilate(img_rgb, tmp3)
```

```python
#Show the images

io.imshow(final1)
```

```python
io.imshow(final2)
```

```python
io.imshow(final3)
```
