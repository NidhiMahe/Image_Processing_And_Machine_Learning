```python
import face_recognition
import cv2
import matplotlib.pyplot as plt

# Load the first image
image1_path = '/users/Nidhi/Desktop/Amitabhb1.jpg'
image1 = face_recognition.load_image_file(image1_path)
image1_face_encodings = face_recognition.face_encodings(image1)

# Load the second image
image2_path = '/users/Nidhi/Desktop/Amitabhb2.jpg'
image2 = face_recognition.load_image_file(image2_path)
image2_face_encodings = face_recognition.face_encodings(image2)

# Check if faces were detected
if len(image1_face_encodings) == 0:
    print("No faces found in the first image.")
if len(image2_face_encodings) == 0:
    print("No faces found in the second image.")

# Compare the faces
if len(image1_face_encodings) > 0 and len(image2_face_encodings) > 0:
    match_results = face_recognition.compare_faces([image1_face_encodings[0]], image2_face_encodings[0])
    if match_results[0]:
        print("The faces are of the same person.")
    else:
        print("The faces are of different people.")

    # Optionally display the images
    image1_bgr = cv2.cvtColor(image1, cv2.COLOR_RGB2BGR)
    image2_bgr = cv2.cvtColor(image2, cv2.COLOR_RGB2BGR)
    
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(image1_bgr, cv2.COLOR_BGR2RGB))
    plt.title('First Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(image2_bgr, cv2.COLOR_BGR2RGB))
    plt.title('Second Image')
    plt.axis('off')

    plt.show()
```
