```python
'''Blending Two Images'''
```

```python
#import required packages
import cv2
```

```python
from skimage import io
```

```python
#Read image 1
img1 = cv2.imread('/users/Nidhi/Desktop/Dog2.jpeg')
```

```python
#Read image 2
img2 = cv2.imread('/users/Nidhi/Desktop/Dog.jpg')
```

```python
#Define alpha and beta
alpha = 0.70
beta = 0.30
```

```python
img2_resized = cv2.resize(img2, (img1.shape[1], img1.shape[0]))
```

```python
#Blend images
final_image = cv2.addWeighted(img1, alpha, img2_resized, beta, 0.0)
```
